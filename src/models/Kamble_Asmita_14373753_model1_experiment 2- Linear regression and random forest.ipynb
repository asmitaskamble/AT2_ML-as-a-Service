```python
# Import merged_data file

import pandas as pd
merged_data = pd.read_csv("merged_data.csv")
```

```python
merged_data
```

```python
X = merged_data.drop(labels=["id","sales"],axis=1)
y = merged_data[["sales"]]
```

```python
X.head()
```

```python
y.head()
```

```python
X.isna().sum()
```

```python
# Split data into cat and con

cat = []
con = []
for i in X.columns:
    if(X[i].dtypes == "object"):
        cat.append(i)
    else:
        con.append(i)
```

```python
con
```

```python
cat
```

```python
for i in con:
    x = X[i].mean()
    X[i]=X[i].fillna(x)

for i in cat:
    x = X[i].mode()[0]
    X[i]=X[i].fillna(x)
```

```python
X
```

```python
X.corr()
```

```python
# Standerdization
import pandas as pd
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X1=pd.DataFrame(ss.fit_transform(X[con]),columns=con)
```

```python
X1
```

```python
from sklearn.preprocessing import LabelEncoder
cat_columns = ['item_id','dept_id','cat_id','store_id','state_id','date','event_name','event_type']
X2 = X[cat_columns].copy()
label_encoders = {}
for column in cat_columns:
    le = LabelEncoder()
    X2[column] = le.fit_transform(X2[column])
    label_encoders[column] = le  
```

```python
for column, le in label_encoders.items():
    print(f"Label Encoding for {column}:")
    for label, original_value in enumerate(le.classes_):
        print(f"Label {label} -> Original Value: {original_value}")
```

```python
X2
```

```python
Xnew=X1.join(X2)
```

```python
Xnew
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Xnew, y, test_size=0.2, random_state=21)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)
```

```python
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'R-squared (RÂ²) Score: {r2}')
```
